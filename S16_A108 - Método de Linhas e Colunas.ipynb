# Renomenado colunas no Pandas

```python
filmes.columns
```

```python
# Maneira errada de se alterar colunas no datatset

filmes.rename(columns = {'star rating':'star_rating','content rating':'content_rating'})
```

A transformação não ocorre diretamente no dataset, porém é possível alterá-lo usando o argumento  'inplace'

```python
#Maneira correta de se alterar o nome das colunas no dataset
filmes.rename(columns = {'star rating':'star_rating','content rating':'content_rating'}, inplace=True)

filmes.columns
```

Podemos também, passar alista de colunas EXPLICITAMENTE

```python
colunas_filmes = ['star rating', 'title', 'content rating', 'genre', 'duration', 'actors list']
filmes.columns = colunas_filmes
filmes.columns
```

Podemos renomear as colunas no momento da leitura do arquivo

# Real World

```python
# Usando funções nativas do Python para renoamer as colunas do dataframe

file='C:\\Users\\nelson.junior\\OneDrive\\Python Scripts\\Curso Uni_dados_Dta_Science\\Movies.csv'
filmes = pd.read_csv(file)
filmes.head()

filmes.columns = filmes.columns.str.replace(' ','_')
filmes.columns
```

# Romovendo colunas do dataframe

```python
# Por boa prática, ao relizarmos mudanças no dataframe, devemos atribuilas a outra variável
# assim, é possível rastrear as alterações e retornar ao estado anterior caso seja necessário.
# usando novamente o argumento 'inplace=True'

# Lembrete Axis = 1 - Colunas / Axis = 0 - Linhas

filmes01 = filmes.drop('content_rating', axis=1)
filmes01.head()
```

```python
# Removendo mais de uma coluna

filmes02 = filmes01.drop(['duration', 'actors_list'], axis=1)
filmes02.head()
```
