```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = []
y = []
y2 = []

for num in range(-200,201,10):
    x.append(num)
    y.append(num)
    y2.append(2 * num)
```

```python
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')
```

# Adicionando descrição aos Eixos

```python
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')

#Primeira descrição
plt.xlabel("Eixo X")
```

# Formatando a fonte descritiva dos eixos

```python
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')

#Primeira descrição
plt.xlabel("Espaço para o eixo X", color = 'red', fontsize = 15)
```

# Alinhamento de textos descritivos

```python
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')

#Primeira descrição
plt.xlabel("Espaço para o eixo X", color = 'red', fontsize = 15, horizontalalignment = 'right')
```

# Y label

```python
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')

#Primeira descrição
plt.xlabel("Espaço para o eixo X", color = 'red', fontsize = 15, horizontalalignment = 'center')
plt.ylabel("Espaço para o eixo Y", color = 'blue', fontsize = 15, horizontalalignment = 'center')
```

### Alinhamento vertical - Alinhamos o eixo y verticalmente através da opção horizontalalignment

```python
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')

#Primeira descrição
plt.xlabel("Espaço para o eixo X", color = 'red', fontsize = 15, horizontalalignment = 'center')
plt.ylabel("Espaço para o eixo Y", color = 'blue', fontsize = 15, horizontalalignment = 'right')
```

### Adicionando título

```python
#Eixos do gráfico
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')

# Descrição
plt.xlabel("Espaço para o eixo X", color = 'red', fontsize = 15, horizontalalignment = 'center')
plt.ylabel("Espaço para o eixo Y", color = 'blue', fontsize = 15, horizontalalignment = 'center')

#Title
plt.title('Título do Gráfico', color ="gray", fontsize = 25)
```

# Texto matemático

```python
#Eixos do gráfico
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue')
plt.plot(x,y2, c = 'green')

# Descrição
plt.xlabel("Espaço para o eixo X", color = 'red', fontsize = 15, horizontalalignment = 'center')
plt.ylabel("Espaço para o eixo Y", color = 'blue', fontsize = 15, horizontalalignment = 'center')

#Title
plt.title(r"Título do Gráfico $\frac{5}{4}$", color ="gray", fontsize = 25) #usando epressão regular "r"
```

# Labels e Legendas

```python
#Eixos do gráfico
plt.plot(x,y, c = 'red', ls ="-", lw = 2, marker ="o", ms = 10, fillstyle = "top", mfc = 'blue', label="Primeiro")
plt.plot(x,y2, c = 'green', label="Segundo") #Inserindo legendas no gráfico

# Descrição
plt.xlabel("Espaço para o eixo X", color = 'red', fontsize = 15, horizontalalignment = 'center')
plt.ylabel("Espaço para o eixo Y", color = 'blue', fontsize = 15, horizontalalignment = 'center')

#Title
plt.title(r"Título do Gráfico $\frac{5}{4}$", color ="gray", fontsize = 25) #usando epressão regular "r"

# Ativando as legendas
plt.legend()
```
